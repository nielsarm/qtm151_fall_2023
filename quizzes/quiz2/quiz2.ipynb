(d) Create a function with four inputs $f(y,x,b0,b1)$

<font size = "3">

$\qquad e2 = (y - b_0 - b_1 * x)^2$

- The function should return the mean of e2
- This is the mean-square error (mse)
- Call the function "fn_mse"
- Let $y$ be the column "Potential"
- Let $x$ be the column "Age" 
- Does <br>
$\qquad (b_0 = 80, b_1 = 0.4)$ or <br>
$\qquad (b_0 = 80, b_1 =-0.3)$ or <br>
$\qquad (b_0 = 80, b_1 = -0.5)$ <br>
produce the lowest value of mse?
- Print your results

````python
# Write your own code
def fn_mse(y,x,b0,b1):
    e2 = (y-b0-b1*x)**2
    return e2.mean()

print(fn_mse(data["Potential"],data["Age"],80,0.4))
print(fn_mse(data["Potential"],data["Age"],80,-0.3))
print(fn_mse(data["Potential"],data["Age"],80,-0.5))

# From the output, we see that (b0 =80, b1 = -0.3) produces the lowest value of MSE. 
````

(e) Split a dataset into subsets

<font size = "3">

- Create ``` list_status = [0,1]```
- Create ``` list_prob = [0.5, 0.5]```
- Create a new variable $n$ with the number of rows of the dataset
- Define a random seed that equals 42

- Use 
```python
    np.random.choice(list_status,
                     size = n,
                     p = list_prob)
```
$\qquad$ to create a new column called "status" <br>
$\qquad$ which randomly assigns each row to "0" or "1" <br>
$\qquad$ with equal probability
- Use .query() to subset all observations with status == 0.<br>
store it in a new dataset called "data0"
- Use .query() to subset all observations with status == 1.<br>
store it in a new dataset called "data1"



````python
# Write your own code
list_status = [0,1]
list_prob = [0.5,0.5]
n = len(data)

np.random.seed(42)
data['status'] = np.random.choice(list_status, n, p=list_prob)
data0 = data.query('status == 0')
data1 = data.query('status == 1')
````

(f) Create two overlapping histogram plots

<font size = "3">

- This question requires the two datasets you created in (e)
- Display the distribution of "Age" for "data0" and "data1"
- Use the "alpha" option to make the graphs semitransparent
- You will need to add a legend, label the axes, and the title

Note: The goal of this question is to illustrate that random <br>
assignment produces very similar distributions between two groups

- Provide a short interpretation of the graph using the information from the Note

````python
# Write your own code
plt.hist(data0['Age'], alpha=0.5, label='status=0')
plt.hist(data1['Age'], alpha=0.5, label='status=1')
plt.legend(loc='upper right')
plt.xlabel('Age')
plt.ylabel('Frequency') 
plt.title('Histogram of Age by Status')
plt.show()
````

From the graph, we can see that much of data for the two histograms overlaps, which suggests that the data is very similar. This is unsurprising as creating the random assignemnt to two different groups from the original dataset can be simply considered a sample. As there was random assignment between the two groups, we can expect that they will look similar as there are more than 9300 observations for each group. 

Additionally, we see that from the original dataset, most soccer players are between 20 and 35 years of age. Also, we note that the mean is likely around 25 years old, as there is a fairly symmetric distribution around that point.
