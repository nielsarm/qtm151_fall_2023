- Display the ```sprint_results_agg``` dataframe
- Provide a one-sentence explanation of the information in the ```sprint_results_agg ``` table

(d) Aggregate and sort

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Obtain the sum of "laps" by "driverId" using ``` .groupby().agg() ```
- Sort the teams in descending order using of the sum of laps using ``` .sort_values() ```
- Display your results

````python
# Write your own code
sprint_results2 = pd.read_csv("/Users/nielsarmbruster/Documents/GitHub/qtm151_fall_2023/quizzes/quiz3/data_raw/sprint_results.csv")
sprint_results2["sum_of_laps"] = sprint_results2.groupby("driverId").agg(sum_laps = ("laps", "sum"))
sprint_results2 = sprint_results2.sort_values(by = ["sum_of_laps"], ascending = False)
display(sprint_results2)
````
