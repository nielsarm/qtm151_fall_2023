```python
# Write your own code
list_age = [58,23,63,18,76,35,48,68]
list_eligible_retire = []
for age in list_age:
    if (age >= 65):
        list_eligible_retire.append(True)
    else:
        list_eligible_retire.append(False)

print(np.sum(list_eligible_retire))
```
